```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('digikala-orders.csv')
data.head()
```

```python
num = data.groupby(['city_name_fa']).size().to_list()
city = np.unique(data['city_name_fa'].to_list())
```

```python
num
```

```python
from bidi.algorithm import get_display
from arabic_reshaper import reshape
city = [get_display(reshape(label)) for label in city]
df = pd.DataFrame({'mass': num}, index=city)
df_up = df[df['mass'] > 5000]
df_low = df.query("mass <= 5000")

sum_df=np.array([df_low]).sum()
new= df_up.append(pd.DataFrame({'mass':sum_df}, index=['other']))
plot = new.plot.pie(y='mass', figsize=(5, 5),autopct='%1.0f%%')
```
